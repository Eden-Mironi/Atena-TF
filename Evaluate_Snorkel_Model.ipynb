# Evaluate Snorkel Model

Evaluate the Snorkel generative model on a gold-labeled test set.

** Requires Snorkel**: `pip install snorkel>=0.9.0`

```python
import numpy as np
import gym
from gym_atena.envs.atena_env_cont import ATENAEnvCont
from gym_atena.envs.atena_snorkel.data_loader import *
from gym_atena.envs.atena_snorkel.snorkel_gen_model import SnorkelGenModel
import Configuration.config as cfg
from Evaluation.notebook_utils import *
import matplotlib.pyplot as plt

%matplotlib inline
```

## Load Test Set

```python
try:
    test_data = load_gold_labels_test_dataset()
    print(f"Loaded {len(test_data)} test examples")
except Exception as e:
    print(f"Could not load test dataset: {e}")
    print("Create one with Create_Snorkel_Testset.ipynb")
    test_data = None
```

## Load Snorkel Model

```python
snorkel_model = SnorkelGenModel()
print(f"Loaded Snorkel model for schema: {cfg.schema}")
```

## Evaluate Model

```python
if test_data:
    predictions = []
    gold_labels = []
    
    for ex in test_data:
        gold_labels.append(ex['gold_humanity_score'])
        pred = snorkel_model.get_coherency_score_prediction(ex['snorkel_data_obj'])
        predictions.append(pred)
    
    predictions = np.array(predictions)
    gold_labels = np.array(gold_labels)
    
    # Calculate metrics
    mae = np.mean(np.abs(predictions - gold_labels))
    print(f"Mean Absolute Error: {mae:.4f}")
```
