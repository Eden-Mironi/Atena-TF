## User Study Session Example

Original session from Flights dataset (3_8)

```python
# Define session actions
action_vecs = [
    [2, 9, 0, 0.0, 0, 0],
    [1, 11, 5, -0.49505794048309326, 0, 0],
    [1, 11, 5, -0.4945875108242035, 0, 0],
    [1, 11, 5, -0.48364895582199097, 0, 0],
    [1, 11, 5, -0.4781721830368042, 0, 0],
    [2, 8, 0, 0.0, 0, 0],
    [1, 11, 5, -0.470523864030838, 0, 0],
    [1, 11, 5, -0.4916938543319702, 0, 0],
    [1, 11, 5, -0.4707660377025604, 0, 0],
    [2, 7, 0, 0.0, 0, 0],
    [1, 11, 5, -0.4091779887676239, 0, 0],
    [1, 11, 5, -0.4697021543979645, 0, 0],
]

filter_terms_lst = [
    None, '25', '18', '1', '17', None,
    '31', '11', '24', None, '26', '10',
]

print(f"Session has {len(action_vecs)} actions")
```

## Run Session

```python
# Run the predefined session
dataset_number = 3  # Flights dataset

info_hist, r_sum = run_episode_analyst_view(
    dataset_number=dataset_number,
    env=env,
    compressed=False,
    filter_by_field=True,
    continuous_filter_term=True,
    actions_lst=action_vecs,
    filter_terms_lst=filter_terms_lst,
)

print(f"\n{'='*60}")
print(f"Total Reward: {r_sum:.2f}")
print(f"Steps Taken: {len(info_hist)}")
print(f"{'='*60}")
```

## Analyze Session Rewards

```python
# Extract reward components
rewards = [info['reward'] for info, _ in info_hist]
cumulative_rewards = np.cumsum(rewards)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Step rewards
axes[0].bar(range(len(rewards)), rewards)
axes[0].set_xlabel('Step')
axes[0].set_ylabel('Reward')
axes[0].set_title('Reward per Step')
axes[0].grid(True, alpha=0.3)

# Cumulative rewards
axes[1].plot(range(len(cumulative_rewards)), cumulative_rewards, marker='o')
axes[1].set_xlabel('Step')
axes[1].set_ylabel('Cumulative Reward')
axes[1].set_title('Cumulative Reward')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
