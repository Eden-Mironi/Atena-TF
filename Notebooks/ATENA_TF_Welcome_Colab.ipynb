## Example 1: Create Environment

```
# Setup
import numpy as np
if not hasattr(np, "bool8"):
    np.bool8 = np.bool_

import gym
import pandas as pd

# Create environment
env = gym.make('ATENAcont-v0')
obs = env.reset(dataset_number=0)

print(f"âœ… Environment created!")
print(f"ðŸ“Š Observation: {obs.shape}")
print(f"ðŸ“‹ Dataset: {env.data.shape}")
display(env.data.head(10))
```

## Example 2: Manual Action

```
# Take a filter action
action = np.array([1, 1, 0, 0.5, 0, 0])
env.render()

obs, reward, done, info = env.step(action)

print(f"Action: {info['action']}")
print(f"Reward: {reward:.2f}")

if info.get('raw_display'):
    df = info['raw_display'][1] or info['raw_display'][0]
    display(df.head(10))
    print(f"New shape: {df.shape}")
```
