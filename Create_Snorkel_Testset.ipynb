# Create Snorkel Test Dataset

Create a gold-labeled test set for evaluating Snorkel labeling functions.

```python
import numpy as np
import gym
from gym_atena.envs.atena_env_cont import ATENAEnvCont
from gym_atena.envs.atena_snorkel.data_loader import (
    save_gold_labels_test_dataset,
    reset_test_dataset    
)
import Configuration.config as cfg
from Evaluation.notebook_utils import *

%matplotlib inline
```

## Setup

```python
env = gym.make(env_d)
print("Environment created")
```

## Reset Test Dataset

```python
reset_test_dataset()
print(" Test dataset reset")
```

## Add Sessions

Define action sequences with humanity scores.

**Scores**: 1.0=human-like, 0.5=neutral, 0.0=not human-like

```python
dataset_number = 0

action_vecs = [
    [2, 3, 1, 0.32, 0, 0],
    [0, 5, 4, 0.6, 0, 0],
    [2, 2, 8, 0.22, 0, 0],
    [2, 6, 8, -0.50, 0, 0],
    [1, 4, 7, -0.47, 0, 0],
]

analyst_humanity_scores = [1.0, 0.0, 0.5, 1.0, 1.0]

add_to_snorkel_testset(env, dataset_number, action_vecs, analyst_humanity_scores)
print(f" Added session with {len(action_vecs)} actions")
```

## Verify Dataset Created

The dataset is automatically saved by `add_to_snorkel_testset()`. Run the cell below to verify.

```python
# Note: The gold labels are already saved by add_to_snorkel_testset()
# Each time you call add_to_snorkel_testset(), it automatically:
# 1. Runs the episode with your actions
# 2. Logs each step to snorkel_test_dataset.jsonl
# 3. Saves the analyst_humanity_scores to snorkel_gold_labels_test_dataset.txt

import os

# Verify the files exist and show their sizes
test_dataset_file = "snorkel_test_dataset.jsonl"
labels_file = "snorkel_gold_labels_test_dataset.txt"

if os.path.exists(test_dataset_file) and os.path.exists(labels_file):
    test_size = os.path.getsize(test_dataset_file)
    labels_count = len(open(labels_file).readlines())
    print(" Test dataset successfully created!")
    print(f"   ðŸ“„ {test_dataset_file}: {test_size} bytes")
    print(f"   ðŸ“„ {labels_file}: {labels_count} labels")
    print(f"\n To add more sessions, create new cells above and call:")
    print(f"   add_to_snorkel_testset(env, dataset_number, action_vecs, analyst_humanity_scores)")
else:
    print("  Dataset files not found. Make sure you ran the add_to_snorkel_testset() cells above.")
```
